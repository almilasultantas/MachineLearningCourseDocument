Polinomal Regresyon
oluşan denklemin derecesinin artması durumunda oluyor 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

veriler = pd.read_csv("maaslar.csv")
#veriler.head()
```

```python
# data frame dilimleme ( slice )
x = veriler.iloc[:,1:2]
y = veriler.iloc[:,2:]
# numpy arayy ( dizi ) dönüşümü
X = x.values
Y = y.values
#print(x)
#print(y)
```

```python
# Linear Regression 
# doğrusal model oluşturma
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, Y)

plt.scatter(X, Y, color = "red")
plt.plot(X,lin_reg.predict(X), color = "blue")
```

```python
# Polynomial regression
# doğrusal olmayan ( nonlinear model ) oluşturma
# 2. dereceden polinom
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(X)
print(x_poly)
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
plt.scatter(X, Y, color = "blue")
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), color = "red")
```

```python
# 4. dereceden polinom
poly_reg3 = PolynomialFeatures(degree = 4)
x_poly3 = poly_reg3.fit_transform(X)
print(x_poly3)
```

```python
lin_reg3 = LinearRegression()
lin_reg3.fit(x_poly3, y)
plt.scatter(X, Y, color = "red")
plt.plot(X, lin_reg3.predict(poly_reg3.fit_transform(X)), color = "blue")
```

Tahminler

```python
print(lin_reg.predict([[6.6]]))
print(lin_reg.predict([[11]]))
```

```python
print(lin_reg2.predict(poly_reg.fit_transform([[6.6]])))
print(lin_reg2.predict(poly_reg.fit_transform([[11]])))
```

```python
print(lin_reg3.predict(poly_reg3.fit_transform([[6.6]])))
print(lin_reg3.predict(poly_reg3.fit_transform([[11]])))
```
